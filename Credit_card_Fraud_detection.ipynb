```python
df=pd.read_csv('creditcard dataset.csv')
df.head()
```

```python
df.describe().T
```

```python
frauds = df.loc[df['Class'] == 1]
frauds.head()
```

# **Balanceamos el Dataset**

```python
no_frauds = df.drop(index = frauds.index)
no_frauds = no_frauds.sample(n = 492)
balanced = pd.concat([no_frauds, frauds])
balanced['Class'].value_counts()
```

```python
balanced
```

```python
fig, ax = plt.subplots(figsize=(20,20))
corr= balanced.corr()
sns.heatmap(corr,ax=ax,annot=True,cmap='BrBG',fmt='.2f',linewidths=0.5,vmin=-1,vmax=1)
```

```python
# Tomamos las columnas que tienen correlacion mas alta
#Tambien puede ser:
#columns = balanced.loc[:,(corr.Class >= 0.48) | (corr.Class <= -0.49)]

columns=balanced.loc[:,(corr['Class']>=0.48)|(corr['Class']<=-0.49)]
columns['Time'] = balanced['Time']
columns['Amount'] = balanced['Amount']
columns
```

```python
balanced = balanced.loc[:, columns.columns]
balanced
```

```python
# Creamos una lista con los modelos que vamos a aplicar en todas las estrategias
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier

classifiers = [DecisionTreeClassifier(),RandomForestClassifier(criterion='entropy',max_depth=3,random_state=0),GradientBoostingClassifier(),AdaBoostClassifier(),XGBClassifier()]
```

# ESTRATEGIA UNDERSAMPLE

```python
X = balanced.drop(columns = 'Class')
y = balanced['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
#Aplicamos cada uno de los modelos
from sklearn.metrics import classification_report,confusion_matrix


for model in classifiers:

  model.fit(X_train, y_train)
  y_pred = model.predict(X_valid)

  print(f'{type(model).__name__}:\n')

  report = classification_report(y_true=y_valid, y_pred=y_pred)
  print(report)

  matrix=confusion_matrix(y_valid,y_pred)
  fig=plt.figure(figsize=(10,5))
  sns.heatmap(matrix,annot=True,fmt='.0f')
  plt.show()


```

# **ESTRATEGIA CON UNDERSAMPLING Y OVERSAMPLING**

```python
from  imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

over = SMOTE(sampling_strategy=0.1)
under = RandomUnderSampler(sampling_strategy=1)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)

df_smote_under = df.loc[:, columns.columns]
X_smo = df_smote_under.drop('Class', axis = 1)
y_smo = df_smote_under['Class']

X_smo, y_smo = pipeline.fit_resample(X_smo, y_smo)
X_train_smo, X_valid_smo, y_train_smo, y_valid_smo = train_test_split(X_smo, y_smo,test_size=0.20, random_state=1)
```

```python
y_smo.value_counts()
```

```python
for model in classifiers:

  model.fit(X_train_smo, y_train_smo)
  y_pred_smo = model.predict(X_valid_smo)

  print(f'{type(model).__name__}:\n')

  report = classification_report(y_valid_smo,y_pred_smo)
  print(report)

  matrix=confusion_matrix(y_valid_smo,y_pred_smo)

  fig=plt.figure(figsize=(10,5))
  sns.heatmap(matrix,annot=True,fmt='.0f')
  plt.show()

```
